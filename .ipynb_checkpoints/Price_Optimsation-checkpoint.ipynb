# ML Project - Price Optimisation



## About Dataset

**Q:** Quantitiy (Miktar)

**BSRSub:** Best Seller Rank (Ilgili Sub Kategori icin)

**Rating:** Musteriler tarafinda 1-5 arasi verilen begeni d*uzeyi

**Review_Numbers:** Urunun aldigi toplam review sayisi

**P:** Price (Fiyat): Urunun ABD'den alis fiyati + OneAMZ ara deponun alacagi handling fee + OneAMZ saticisinin tahmini %15 civarindaki kari + Amazon'un nihai fiyat uzerinden alacagi ortalama %15 fee 

**R:** Revenue (Gelir): P - (Urunun ABD'den alis fiyati + OneAMZ ara deponun alacagi handling fee+ Amazon'un nihai fiyat uzerinden alacagi ortalama %15 fee)

**QP:** Miktrar ile fiyatin carpimi (Q*P) olup, ciroyu verir.

**QP2:** Q*P'nin karesi: Ters U seklinde parabolik bir fonksiyonla calismak istedigimiz icin temel aciklayici degiskenimiz olan QP'nin 2.derece degerini de olusturup modele ayri bir degiskenmis gibi ekliyoruz. 



**Model:** R=β0+β1*QP + β2*QP2+ β3*BSRSub+β4*SellerNumbers+β5*Rating + β67*ReviewNumbers

Bu modelde asil belirleyici olan kisim:  **R=β0+β1*QP + β2*QP2** bolumudur. Bu kisimla asagidaki sekle benzer bir goruntuye ulasilmaya calisilmaktadir. 



Bu model tahmin edildikten sonra; Q*P'ye gore kismi turevi alinip, sifira esitlenip, Q*P degeri (optimum ciro) bulunacaktir. Sonra bu ciro, urun sayisina bolunerek, optimum fiyat (price) degerine ulasilacaktir. Devaminda arzu edilirse OneAMZ saticisinin nihai net kar orani (%) da belirlenebilmektedir. 




**Not:** Neden burada 2. dereceden polinominal foksiyon kullanilmadigi sorusu akla gelebilir. O durumda yontem tarafindan tum degiskenlerin kareleri alinmakta, ikili, uclu, dortlu carpimlari da isin icine katilmaktadir. Oysa bizim ihtiyacimiz olan; temel aciklayici degiskenle kurulacak ters U seklindeki parabolik bir fonksiyon ve bu fonksiyonda ommitted variable (eksik degisken kullanimi) sorunu olusmasini onleyebilmek ve bu yolla  β1 ve β2'nin gercek degerlerini ortaya cikarabimek icin diger aciklayici degiskenleri (features) modele dahil etmektir.

# ML

**ML basamaklari (From Ms Lilibet):**

**1.** En basta targetimi y, featureslari da X olarak atiyorum. Sonrasinda Train ve test olarak datami boluyorum.(train_test_split) (min %70 train, %30test olacak)

**2.** Elimde X_train, y_train, X_test, y_test datasi olacak.

**3.** Model’i linear regression olarak tanimliyorum.(model=LinearRegression()). coefficient ve intercepti cagirarak katsayilari bulduruyorum

**4.** X-train ile y_train’i model fit kullanarak egitiyorum (model.fit)

**5.** X_test datasetini kullanarak bir y-predict elde ediyorum (model.predict)
Bunu yaparken elde ettigim regresson line’ina (y=aX1+bX2+cX3+dX4+e) X_test verilerini vererek y_predict elde etmis oluyorum. Burdaki a,b,c,d,e’yi de coefficient ve interceptleri cagirarak bulmustum.

**6.** Sonra y-predict ile y-actual’i, yani y-testi kiyasliyorum. Residual’lari elde ediyorum. Bakiyorum hata payim ne kadar diye. Bunu da eval_metric ile error metricleri cagirarak yapiyorum.

**7.** Train datasetimin error metriclerini de cagirip, test sonucunda elde ettiklerim ile kiyasliyorum.

**8.** Boyle boyle denedigim modeller icinde en iyi sonuc aldigim modele, final_model olarak tum datami sokuyorum.

## Scaling the Data

Train datasına fit yapip, train ve test datalarına transform yapacagiz.

```python
#Scaling isleminde kullanacagimiz kutuphaneleri cagiralim;

from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
scaler = StandardScaler() # burada 3 secenek var, biz standardizasyon yapacagimiz icin bunu secik
                          # MinMaxScaler; normalizasyon icin kullanilacak
```

```python
# fit'de "X_train" datasi uzerinden nasil scale (standardize) edecegini ogreniyor

scaler.fit(X_train)  
```

```python
# Bir ustte ogrendigi (belirledigi) olceklendirme sistemini "transform"la X_train datasina uyguluyor

X_train_scaled = scaler.transform(X_train)  
X_train_scaled 
```

!!! **NOT:** "**X_train**" datasina **hem fit, hem de transform** uygularken, "**X_test**" datasina sadece transform uygulanir. 

!!! **y_train** ve **y_test** datalarina ise **scaling** islemleri **uygulanmaz!!!!!!!!** 

```python
# X_test serisini scale edebilmek icin sadece transform yapacagiz. 
# Boylece X_test'i scale ederken "fit" ile ogrendigi bigilerle X_test serisini de scale eder.

X_test_scaled = scaler.transform(X_test) 
X_test_scaled 
```

* Ham halleri buyuk sayilari da iceren X_train ve X_test serileri "standardize" edilince degerleri -1, +1 araliginda degisir hale geldi.

* Standardize edilmis serilerin ortalamasi 0, standart sapmasi 1'dir. Degerleri -2, +2 de oabilir illaki -1, +1 arasinda olaak diye bir sart yok. Ama kuculuyor degerleri. BU calismada 8, 9 gibi degerler de var scale edilmis test datasinda.

* Artik elimizde **scale edilmis X_train_scaled ve X_test_scaled verileri var.**

###  eval_metrics for "Train Data"

```python
# Eval metrics icin kullanacagimiz kutuphaneyi cagiralim;

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Train Datasi icin eval metricsleri elde edelim; 

def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    #print("Model testing performance:")
    print("---------------------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
    
print("Model testing performance for TRAIN DATA:")   



eval_metric(y_train, y_train_pred)
```

### Elde Ettigimiz Modeli "Test" Verilerine Uygulayalim


* Elde ettigimiz modeli "Test" datalarini kullanarak sinayalim. 

* Bunun icin X_test verilerini elde ettigimiz modeldeki X yerine yazalim;

```python
y_pred = lin_reg.predict(X_test_scaled)
y_pred
```

### eval_metrics for "Test Data"

```python
# Test Datasi icin eval metricsleri elde edelim; 

def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    #print("Model testing performance:")
    print("----------------------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
    
    
print("Model testing performance for TEST DATA:")   


eval_metric(y_test, y_pred)
```

### Train ve Test datalari icin yaptigimiz tahmin sonuclarini "eval_metrics" uzerinden karsilastirmali degerlendirelim; 

```python
print("Linear Model testing performance for TRAIN DATA:")
print(eval_metric(y_train, y_train_pred))

print("---------------------------------------")

print("Linear Model testing performance for TEST DATA:")
print(eval_metric(y_test, y_pred))
```

### Multiple Linear 'a ait eval metrikler icin 2. Yol: 

```python
def train_val(y_train, y_train_pred, y_test, y_pred, i):
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
linear = train_val(y_train, y_train_pred, y_test, y_pred, "Linear")
linear
```
